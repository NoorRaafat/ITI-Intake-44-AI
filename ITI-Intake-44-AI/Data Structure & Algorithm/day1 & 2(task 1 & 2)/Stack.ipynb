```python
class Node:
    def __init__(self,data):
        self.data=data
        self.prev=None
```

```python
class Stack:
    def __init__(self):
        self.tail=None
        # push in the end 
    def push(self,stud_id,Name):
        node = Node((stud_id, Name))
        
        if self.tail is None:
            self.tail=node
        else:
            node.prev=self.tail
            self.tail=node
    def pop(self):
        # to delete from end (علي طول لاني مبيكنش عندي اصلا tail) 
        node=None
        if self.tail is not None:
            node=self.tail
            self.tail=self.tail.prev
        return node
    
    def Display_Node_all(self):
        cur=self.tail
        while(cur is not None):
            print("student_ID:", cur.data[0], "Name:", cur.data[1])
            print(cur.data)
            cur=cur.prev   
    def exit(self):
        print("Exiting program...")
        
    def free(self):
        current=self.tail
        while current:
            self.pop()
            current=current.prev
        print('stack is now free')
        
```

```python
def menu():
    stack = Stack()
    while True:
        print("\nStack Menu:")
        print("1. Push")
        print("2. Pop")
        print("3. Display All Nodes")
        print("4. Free Stack")
        print("5. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == '1':
            stud_id = input("Enter student ID: ")
            name = input("Enter student name: ")
            stack.push(stud_id, name)
            print("Node pushed onto stack.")
        elif choice == '2':
            popped_node = stack.pop()
            if popped_node:
                print("Popped node - student ID:", popped_node.data[0], "Name:", popped_node.data[1])
            else:
                print("Stack is empty. Nothing to pop.")
        elif choice == '3':
            print("\nDisplaying all nodes in stack:")
            stack.Display_Node_all()
        elif choice == '4':
            stack.free()
        elif choice == '5':
            stack.exit()
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```

```python

menu()
```
