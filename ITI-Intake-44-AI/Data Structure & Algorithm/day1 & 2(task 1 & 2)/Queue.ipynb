```python
class Queue:
    def __init__(self):
        self.queue = []
        self.max_size = 5

    def enqueue(self, stud_id, Name):
        if len(self.queue) < self.max_size:
            self.queue.append((stud_id, Name))
        else:
            print("Queue is full. Cannot enqueue more elements.")

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        else:
            print("Queue is empty")
            return None

    def display_all(self):
        for stud_id, Name in self.queue:
            print("student_ID:", stud_id, "Name:", Name)

    def is_empty(self):
        return len(self.queue) == 0

    def exit(self):
        print("Exiting program...")

    def free(self):
        self.queue = []
        print('Queue is now free')
```

```python
def main():
    queue = Queue()
    while True:
        print("\nQueue Menu:")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Display all students")
        print("4. Check if queue is empty")
        print("5. Free the queue")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            stud_id = input("Enter student ID: ")
            name = input("Enter student Name: ")
            queue.enqueue(stud_id, name)
            print("Student added to the queue.")

        elif choice == "2":
            student = queue.dequeue()
            if student:
                print("Dequeued Student ID:", student[0], "Name:", student[1])

        elif choice == "3":
            queue.display_all()

        elif choice == "4":
            if queue.is_empty():
                print("Queue is empty.")
            else:
                print("Queue is not empty.")

        elif choice == "5":
            queue.free()
        
        elif choice == "6":
            queue.exit()
            break

        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
