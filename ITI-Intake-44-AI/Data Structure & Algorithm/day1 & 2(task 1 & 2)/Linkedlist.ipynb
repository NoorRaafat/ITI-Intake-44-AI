```python
def menu():
    doubly_linked_list = Double_linkedlist()
    while True:
        print("\n---  Menu ---")
        print("1. Add a student")
        print("2. Insert a student at a specific position")
        print("3. Search for a student by name")
        print("4. Delete a student by ID")
        print("5. Display all students")
        print("6. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            stud_id = input("Enter student ID: ")
            name = input("Enter student name: ")
            doubly_linked_list.add(stud_id, name)
            print("Student added successfully!")
            
        elif choice == '2':
            loc = int(input("Enter position to insert: "))
            stud_id = input("Enter student ID: ")
            name = input("Enter student name: ")
            doubly_linked_list.insert(loc, stud_id, name)
            print("Student inserted successfully!")
        
        elif choice == '3':
            name = input("Enter student name to search: ")
            result = doubly_linked_list.Search(name)
            if result:
                print("Student found - ID:", result.data[0], "Name:", result.data[1])
            else:
                print("Student not found.")
        
        elif choice == '4':
            stud_id = input("Enter student ID to delete: ")
            if doubly_linked_list.delete(stud_id):
                print("Student deleted successfully!")
            else:
                print("Student not found.")
        
        elif choice == '5':
            print("\n--- All Students ---")
            doubly_linked_list.Display_Node_all()
        
        elif choice == '6':
            doubly_linked_list.exit()
            break
        
        else:
            print("Invalid choice! Please enter a valid option.")
```
